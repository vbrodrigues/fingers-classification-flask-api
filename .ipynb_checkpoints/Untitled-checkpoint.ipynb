```python
from PIL import Image
import base64
import cv2
import io
import numpy as np
import uuid
```

```python
img = Image.open("0.png")
```

```python
img.getcolors
```

```python
img.convert("RGB").size
```

```python
img_bytes = img.tobytes()
img_bytes
```

```python
img_b64 = base64.b64encode(img_bytes)
img_b64
```

```python
base64.b64decode(img_b64)
```

```python
from_b64 = base64.b64decode(img_b64)
```

```python
from_b64
```

```python
Image.frombytes("L", (128, 128), from_b64)
```

```python
img = cv2.imread("0.png")
```

```python
img_bytes = img.tobytes()
```

```python
img_b64 = base64.b64encode(img_bytes)
img_b64
```

```python
from_b64 = base64.b64decode(str(img_b64)[2:-1].encode())
```

```python
img_pil = Image.frombytes("RGB", (128, 128), from_b64)
```

```python
img_pil.save("test.png", format="png")
```
